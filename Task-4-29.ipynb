Велосипедист должен проехать путь из пункта A в B в определённый срок. Если он будет ехать со скоростью 12
км / ч, то опоздает на 0,5 часа; если же поедет со скоростью 15 км/ч, то приедет в пункт B на 12 мин. раньше
срока. Определить расстояние между пунктами A и B.

```python
from sympy.abc import s,v
from sympy import Eq
from sympy import solve

sol = solve([Eq(s/12-s/v, 1/2), Eq(s / v - s / 15, 1/5)], [s,v])
ab = sol[0][0]
print(sol)
print("AB Length:",ab)
```
